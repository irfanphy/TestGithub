```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def take_data(filename):
    data = np.loadtxt(filename)
    x, y, rxx, rh = data[:, 0], data[:, 1], data[:, 3], data[:, 4]
    nx = 0
    ny = 0
    aux = y[0]
    n=len(x)
    for i in range(n):
        for j in range(n):
            y[j] != aux
            ny = ny + 1
            aux =y[j]
            nx = 0
        nx = nx+1
    return nx, ny
```

```python
filename = 'landaufan.dat'
```

```python
nx, ny =take_data(filename)
```

```python
data = np.loadtxt(filename)
Rxx=data[:,3]
```

```python
W=Rxx.reshape(401, 81)
```

```python
plt.imshow(W)
plt.show()
```
