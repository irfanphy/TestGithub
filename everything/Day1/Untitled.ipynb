```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def take_data(filename):
    data=np.loadtxt(filename)
    x, y, rxx, rh = data[:, 0], data[:, 1], data[:, 3], data[:, 4]
    
    aux = y[0]
    nx = 0
    ny = 1
    for i in range(len(x)):
        if y[i] != aux:
            ny = ny + 1
            nx = 0
        aux=y[i]    
        nx =nx + 1
    Rxx = rxx.reshape(ny, nx)
    RH = rh.reshape(ny, nx)
    X = x
    Y = y
    
    return X, Y, Rxx, RH
    
    
```

```python
def plot_data(X, Y, Rxx, RH):
    fig = plt.figure(figsize=(16, 8.5))
    ax1 = fig.add_subplot(2, 2, 1)
    im1 = ax1.imshow(Rxx, aspect='auto', origin='lower', extent=[min(X), max(X), min(Y), max(Y)])
    ax1.set_xlabel('$X$')
    ax1.set_ylabel('$Y$')
    ax1.set_title('Rxx')
    plt.colorbar(im1)
    
    ax2 = fig.add_subplot(2, 2, 2)
    im2 = ax2.imshow(RH, aspect='auto', origin='lower', extent=[min(X), max(X), min(Y), max(Y)])
    ax2.set_xlabel('$X$')
    ax2.set_ylabel('$Y$')
    ax2.set_title('RH')
    plt.colorbar(im2)
    plt.show()
```

```python
filename = 'landaufan.dat'
X, Y, Rxx, RH = take_data(filename)
plot_data(X, Y, Rxx, RH)
```
