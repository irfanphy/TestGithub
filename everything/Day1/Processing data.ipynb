```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def take_data(filename):
    data = np.loadtxt(filename)
    x, y, rxx, rh = data[:, 0], data[:, 1], data[:, 3], data[:, 4]
    nx = 0
    ny = 0
    aux = y[0]
    n=len(x)
    for i in range(n):
        if y[i] != aux:
            ny = ny + 1
            aux =y[i]
            nx = 0
        nx = nx+1
    return nx, ny
```
